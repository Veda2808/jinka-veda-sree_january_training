2) Implement 2 examples each of " For " loop and " While " loop using python

```python
# Print numbers from 1 to 5
for i in range(1, 6):
    print(i)
```

```python
# Print elements of a list
colors = ["Red", "Green", "Blue"]
for color in colors:
    print(color)
```

```python
# Print numbers from 1 to 5
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
# Countdown from 5 to 1
count = 5
while count > 0:
    print(count)
    count -= 1
```
